```julia
x = 100*rand(50)
y = 100*rand(50)
areas = 800*rand(50)

#Plot#
fig = figure("pyplot_scatterplot",figsize=(8,4))
ax = axes()
scatter(x,y,s=areas,alpha=0.5,color="blue")

title("Scatter Plot")
xlabel("X")
ylabel("Y")
grid("on")
```

```julia
labels = ["Imperial","Stella","Quilmes","Isenbeck"]
colors = ["yellowgreen","gold","lightskyblue","lightcoral"]
sizes = NaN*zeros(4)
explode = zeros(length(sizes))
explode[round(length(sizes)*rand(1))] = 0.1
sizes = [15, 30, 45, 10]

#Plot#
fig = figure("pyplot_piechart",figsize=(5,5))
p = pie(sizes,
		labels=labels,
		shadow=true,
		startangle=90,
		explode=explode,
		colors=colors,
		autopct="%1.1f%%")

axis("equal")
title("Cerveza")
```

```julia
n = 100
x = linspace(-3, 3, n)
y = linspace(-3,3,n)
xgrid = repmat(x',n,1)
ygrid = repmat(y,1,n)
z = zeros(n,n)
for i in 1:n
    for j in 1:n
        z[i:i,j:j] = pdf(MvNormal([eye(2)]),[x[i] y[j]]')
    end
end

#Plot#
fig = figure("pyplot_surfaceplot",figsize=(8,8))
ax = fig[:add_subplot](2,1,1, projection = "3d") 
ax[:plot_surface](xgrid, ygrid, z, rstride=2,edgecolors="k", cstride=2, cmap=ColorMap("gray"), alpha=0.8, linewidth=0.25) 
xlabel("X") 
ylabel("Y")
title("Surface Plot")
```

```julia
#################
#  Create Data  #
#################
x = [1:1:100]
y = ones(100)
for i=70:1:100
    y[i] = y[i] - (0.7/30)*(i-70)
end

##############
#  XKCD Plot #
##############
xkcd() # Set to XKCD mode, based on the comic (hand drawn)
fig = figure("pyplot_inexact",figsize=(8,4))
ax = axes()
p = plot(x,y)
ax[:set_ylim]([0.2,1.1])
annotate("El da que\ndescubr Julia...",xy=[70,1],arrowprops=["arrowstyle"=>"->"],xytext=[25,0.8])

xticks([])
yticks([])
xlabel("Das")
ylabel("Salud")
title("Plot con formato xkcd")

ax[:spines]["top"][:set_color]("none") # Remove the top axis boundary
ax[:spines]["right"][:set_color]("none") # Remove the right axis boundary
fig[:canvas][:draw]() # Update the figure
```
