Gadfly es el paquete de gráficos nativo de Julia. Todavía es bastante joven...

```julia
using Gadfly
using DataFrames
```

```julia
x=[1:100.0];
```

```julia
for i = 1:length(x)
    if isodd(i) == true
        x[i] += x[i]*rand()
    else
        x[i] -= x[i]*rand()
    end
end
```

```julia
plot(x=1:length(x), y=x)
```

```julia
plot(x=1:length(x), y=x, Geom.line)
```

```julia
plot(x=1:length(x), y=x, Geom.point, Geom.smooth)
```

```julia
plot(x=1:length(x), y=x, Geom.bar)
```

en realidad está hecho para usar DataFrames

```julia
xs = 0:0.1:20
```

```julia
df_cos = DataFrame(
x=xs,
y=cos(xs),
ymin=cos(xs) .- 0.5,
ymax=cos(xs) .+ 0.5,
f="cos")
```

```julia
plot (df_cos, x=:x, y=:y, ymin=:ymin, ymax=:ymax, color=:f, Geom.line, Geom.ribbon)
```
